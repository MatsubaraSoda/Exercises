```python
# 16.61
f = 784 # Hz
v = 344 # m/s
λ = v / f # s

x = sp.symbols('x', positive = True)
AB = 2
BC = x
AC = sp.sqrt(BC**2 + AB**2)

L = AC - BC
n = sp.symbols('n', integer = True)
eq = sp.Eq(L, n * (λ/2))
sol = sp.solve(eq, x)

a = [sol[0].subs(n, i).evalf(3) for i in range(1,9,2)]
print(f"(a) {a}.")

eq = sp.Eq(L, n * λ)
sol = sp.solve(eq, x)

b = [sol[0].subs(n, i).evalf(3) for i in range(1,5,1)]
print(f"(b) {b}.")

f_max = v / (2 * AB)
print(f"(c) {f_max:.3g} Hz.")
```
