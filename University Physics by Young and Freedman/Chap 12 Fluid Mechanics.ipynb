```python
# 12.77 CP
H, h, g= sp.symbols('H h g', positive=True)
v = sp.sqrt(2 * g * h)
t = sp.sqrt(2 * (H - h) / g)
R = v * t
display(R.nsimplify())
print()

x = sp.Symbol('h_2', positive=True)
eq = sp.Eq(R.subs(h, H - x), R)
sol = sp.solve(eq, x)
display(sol)
```
